## standardise features

```python
def clean_price_column(df, col):
    """
    Removes dollar signs, commas, whitespace, and converts to float.
    Invalid values become NaN.
    """
    df[col] = (
        df[col]
        .astype(str)
        .str.replace("$", "", regex=False)
        .str.replace(",", "", regex=False)
        .str.strip()
    )
    df[col] = pd.to_numeric(df[col], errors="coerce")
    return df
```

```python
def extract_brand(product_name):
    """
    Extract brand from the beginning of a product name.

    Rules:
    - Capture the first capitalized word(s)
    - Stop when:
        * a word starts with a digit
        * lowercase word appears (means description, not brand)
        * token looks like a model number (letters+digits)
    - If first word has hyphen (e.g., 'DB-Longboards'), take left side.
    """

    if not isinstance(product_name, str):
        return np.nan

    name = product_name.strip()
    tokens = name.split()

    # Handle hyphenated first token → likely brand prefix
    first = tokens[0]
    if "-" in first:
        left = first.split("-")[0]
        if left.isalpha() and left[0].isupper():
            return left

    brand_tokens = []
    for token in tokens:
        # Stop at numbers
        if re.match(r"^\d", token):
            break

        # Stop if mixed letter+digit like 'X500', 'AB12'
        if re.match(r"^[A-Za-z]+\d+", token):
            break

        # Stop if lowercase token appears
        if token[0].islower():
            break

        # Accept this token as part of the brand
        brand_tokens.append(token)

    if not brand_tokens:
        return np.nan

    return " ".join(brand_tokens)
```

```python
def extract_main_category(val):
    """Extract the main category before the first ' | ' delimiter."""

    if pd.isna(val):
        return None

    s = str(val).strip()

    # Split at " | "
    main_cat = s.split(" | ")[0].strip()

    return main_cat.lower()
```

```python
def extract_weights(val):
    """
    Extract a weight from a string and convert everything to pounds.

    Examples:
    - '16 ounces' → 1.0 lb
    - '32 oz' → 2.0 lb
    - '2 pounds' → 2.0 lb
    - '1.5 lbs' → 1.5 lb
    - '0.25 pound' → 0.25 lb
    """

    if pd.isna(val):
        return np.nan

    s = str(val).strip().replace(",", "").lower()

    # Regex captures:
    #  - number with possible decimal
    #  - unit (pounds, pound, lbs, lb, ounces, ounce, oz)
    match = re.search(r"([0-9]*\.?[0-9]+)\s*(pounds|pound|lbs|lb|ounces|ounce|oz)", s)
    if not match:
        return np.nan

    num = float(match.group(1))
    unit = match.group(2)

    # Convert to pounds
    if unit in ["ounces", "ounce", "oz"]:
        return num / 16.0  # 16 ounces = 1 pound

    return num  # already pounds
```

```python
def extract_price_per(price, weight):
    """Computes price per unit (price divided by extracted weight)."""

    # Return NaN if either value is missing
    if pd.isna(price) or pd.isna(weight):
        return np.nan

    return price / weight
```

```python
merged_clean = clean_price_column(merged_clean, "selling_price")
merged_clean.rename(columns={"selling_price": "price"}, inplace=True)
```

```python
merged_clean["brand"] = merged_clean["product_name"].apply(extract_brand)
```

```python
merged_clean["main_category"] = merged_clean["category"].apply(extract_main_category)
```

```python
merged_clean["weight_lb"] = merged_clean["shipping_weight"].apply(extract_weights)
```

```python
merged_clean["price_per_lb"] = merged_clean.apply(
    lambda row: extract_price_per(row["price"], row["weight_lb"]),
    axis=1
)
```
